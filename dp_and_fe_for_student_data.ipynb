```python
import pandas as pd
```

```python
df = pd.read_csv('student.csv')
df.head()
```

```python
df.loc[df['GPA'] < 0, 'GPA'] = None
df.loc[(df['อายุ'] < 15) | (df['อายุ'] > 100), 'อายุ'] = None
```

GPA ไม่ควรจะติดลบ และอายุผิดปกติสูงเกินจริง จะเป็น NaN

```python
gender_map = {
    'ชาย': 'ชาย', 'ช': 'ชาย', 'Male': 'ชาย',
    'หญิง': 'หญิง', 'ญ': 'หญิง', 'Female': 'หญิง'
}
```

```python
df['เพศ'] = df['เพศ'].map(gender_map)
```

เพื่อไม่ให้สับสนเพราะเขียนต่างกันแต่ความหมายเดียวกัน

```python
faculty_map = {
    'วิทยาศาสตร์': 'วิทยาศาสตร์', 'วิทย์': 'วิทยาศาสตร์', 'Science': 'วิทยาศาสตร์', 'คณะวิทยาศาสตร์': 'วิทยาศาสตร์',
    'วิศวกรรมศาสตร์': 'วิศวกรรมศาสตร์', 'วิศวะ': 'วิศวกรรมศาสตร์', 'Engineering': 'วิศวกรรมศาสตร์'
}
```

```python
df['คณะ'] = df['คณะ'].map(faculty_map)
```

ลดความสับสนจากชื่อคณะ เพื่อที่จะทำสถิติได้ถูกต้อง เพราะอาจจะนับคณะซ้ำเพราะเขียนต่างกัน

```python
num_cols = ['อายุ', 'GPA', 'ส่วนสูง_cm', 'น้ำหนัก_kg', 'ชั่วโมงเรียน_ต่อสัปดาห์']
```

```python
for col in num_cols:
   df[col] = df[col].fillna(df[col].mean())
```

ใช้ค่าเฉลี่ยเพื่อดูแนวโน้มของข้อมูล

```python
cat_cols = ['ชื่อ', 'เพศ', 'คณะ']
```

```python
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```
